```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
ax.grid()
```

```python
# Create the plot and add a label for the legend.
#add color, marker points, line width
plt.plot(x_axis, y_axis, marker="*", label='Boston', color="green", linewidth=2)
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# create the plot, vertical bar chart
plt.bar(x_axis, y_axis)
```

```python
# create plot
plt.bar(x_axis, y_axis, color="green", label = "Boston")
#create the labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($))")
# create a title
plt.title("Pyber Fare by Month")
# add the legend
plt.legend()
```

```python
# turn the above graph to horizontal, create the plot
plt.barh(x_axis, y_axis)
```

```python
# switch the x and y axis
plt.barh(y_axis, x_axis)
```

```python
# arrange the y and x axis back to normal and put y-axis in acending order
plt.barh(x_axis, y_axis, color ="magenta", label ="Boston")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()

plt.legend()

# gac is get current axis
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# bubble chart adds third factor which is size. When you have more than 1 dataset using same axis
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# make y_axis more noticable
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# use list comprehension to replace for and while loops, faster method
#new_list = [expression FOR item IN LIST IF conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# create scatter plot using objext-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# create pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# add percentages to the months and "epxlode" largest value. largest value = July, which is 7th value
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# adding colors to pie chart. We have to assign colors to an array 
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```
